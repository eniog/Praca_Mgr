```python
# Na start zainstaluj na dysku google usługe google colaboratory
# Następnie pobierz pliki test.csv i train.csv z https://github.com/eniog/Praca_Mgr.git do Katalogu Praca_Mgr na dysk google
from google.colab import drive # Importuj dane z dysku do Google Colaboratory,akceptuj warunki,wklej kod aktywacyjny
drive.mount('/content/drive')
import pandas as pd # alias blioteki pandas
df_train = pd.read_csv('/content/drive/MyDrive/Praca_Mgr/train.csv') # pobieramy dane do testów
df_test = pd.read_csv('/content/drive/MyDrive/Praca_Mgr/test.csv') # pobieramy dane do testów
df_train.head(10) # sprawdzanie czy dane pobrane prawidłowo
df_train.info() # 891 id, 889 Embarked, 714 age, 204 Cabin dane niekompletne
print('train',df_train.shape) # rozmiar tablicy
print('test',df_test.shape) # brak kolumny Survived, którą należy przewidzieć
print('suma',len(df_train) + len(df_test)) # ilość wszystkich rekordów
#pierwszy test jaki wpływ ma na przeżycie klasa i płeć
print(df_train.groupby(['Pclass','Sex'])['Survived'].value_counts(normalize=True))

#porównanie danych w tabelach train i test na przyładzie
describe_field = ['Age','Fare','Pclass','SibSp','Parch']
print('Train: males')
print(df_train[df_train.Sex =="male"][describe_field].describe())

print('Test: males')
print(df_test[df_test.Sex =="male"][describe_field].describe())

print('Train: females')
print(df_train[df_train.Sex =="female"][describe_field].describe())

print('Test: females')
print(df_test[df_test.Sex =="female"][describe_field].describe())

# pierwsze testy z prostym klasyfikatorem
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

X = df_train[['Fare']].values
y = df_train['Survived'].values
# tworzymy model 
model = DummyClassifier()
# trenujemy 
model.fit(X, y)
# przewidujemy
y_pred = model.predict(X)

score = accuracy_score(y, y_pred)
print("score: %.2f" % score)
# Predykcja danych
X = df_test[['Fare']].values

df_test['Survived'] = model.predict(X)
df_test[ ['PassengerId', 'Survived'] ].to_csv('/content/drive/MyDrive/Praca_Mgr/dummy.csv', index=False)
# pobranie i odczytanie stworzonego pliku
df_dummy = pd.read_csv('/content/drive/MyDrive/Praca_Mgr/dummy.csv')
print(df_dummy.head(10))
print("koniec na dzisiaj, trafione przewidywanie na 0.55")
print("Jesli się połączycie to dodajcie na końcu swojego printa")


```
